````

public interface IAnimal
{
    string Name { get; set; }           
    int Age { get; set; }               
    string MakeSound();                 
    string Move();                      
    string GetInfo();                   
}

public class Dog : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string Breed { get; set; }   

    public Dog(string name, int age, string breed)
    {
        Name = name;
        Age = age;
        Breed = breed;
    }

    public string MakeSound()
    {
        return "–ì–∞–≤-–≥–∞–≤!";
    }

    public string Move()
    {
        return "–ë–µ–∂–∏—Ç –Ω–∞ —á–µ—Ç—ã—Ä–µ—Ö –ª–∞–ø–∞—Ö";
    }

    public string GetInfo()
    {
        return $"–°–æ–±–∞–∫–∞: {Name}, –í–æ–∑—Ä–∞—Å—Ç: {Age} –ª–µ—Ç, –ü–æ—Ä–æ–¥–∞: {Breed}";
    }
}

public class Cat : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public bool IsIndoor { get; set; }  

    public Cat(string name, int age, bool isIndoor)
    {
        Name = name;
        Age = age;
        IsIndoor = isIndoor;
    }

    public string MakeSound()
    {
        return "–ú—è—É-–º—è—É!";
    }

    public string Move()
    {
        return "–ö—Ä–∞–¥–µ—Ç—Å—è –±–µ—Å—à—É–º–Ω–æ";
    }

    public string GetInfo()
    {
        string type = IsIndoor ? "–¥–æ–º–∞—à–Ω—è—è" : "—É–ª–∏—á–Ω–∞—è";
        return $"–ö–æ—à–∫–∞: {Name}, –í–æ–∑—Ä–∞—Å—Ç: {Age} –ª–µ—Ç, –¢–∏–ø: {type}";
    }
}

public class Bird : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public double Wingspan { get; set; } 

    public Bird(string name, int age, double wingspan)
    {
        Name = name;
        Age = age;
        Wingspan = wingspan;
    }

    public string MakeSound()
    {
        return "–ß–∏–∫-—á–∏—Ä–∏–∫!";
    }

    public string Move()
    {
        return "–õ–µ—Ç–∞–µ—Ç –≤ –Ω–µ–±–µ";
    }

    public string GetInfo()
    {
        return $"–ü—Ç–∏—Ü–∞: {Name}, –í–æ–∑—Ä–∞—Å—Ç: {Age} –ª–µ—Ç, –†–∞–∑–º–∞—Ö –∫—Ä—ã–ª—å–µ–≤: {Wingspan} —Å–º";
    }
}

public class Fish : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string WaterType { get; set; } 

    public Fish(string name, int age, string waterType)
    {
        Name = name;
        Age = age;
        WaterType = waterType;
    }

    public string MakeSound()
    {
        return "–ë—É–ª—å-–±—É–ª—å!";
    }

    public string Move()
    {
        return "–ü–ª–∞–≤–∞–µ—Ç –≤ –≤–æ–¥–µ";
    }

    public string GetInfo()
    {
        return $"–†—ã–±–∞: {Name}, –í–æ–∑—Ä–∞—Å—Ç: {Age} –ª–µ—Ç, –¢–∏–ø –≤–æ–¥—ã: {WaterType}";
    }
}

public class AnimalService
{
    private readonly IAnimal _animal;

    public AnimalService(IAnimal animal)
    {
        _animal = animal;
    }

    public void DisplayAnimalInfo()
    {
        Console.WriteLine("=== –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∂–∏–≤–æ—Ç–Ω–æ–º ===");
        Console.WriteLine(_animal.GetInfo());
        Console.WriteLine($"–ó–≤—É–∫: {_animal.MakeSound()}");
        Console.WriteLine($"–î–≤–∏–∂–µ–Ω–∏–µ: {_animal.Move()}");
        Console.WriteLine();
    }

    public void PerformDailyRoutine()
    {
        Console.WriteLine($"--- –ï–∂–µ–¥–Ω–µ–≤–Ω–∞—è —Ä—É—Ç–∏–Ω–∞ –¥–ª—è {_animal.Name} ---");
        Console.WriteLine($"1. {_animal.Move()}");
        Console.WriteLine($"2. {_animal.MakeSound()}");
        Console.WriteLine($"3. {_animal.Name} –æ—Ç–¥—ã—Ö–∞–µ—Ç");
        Console.WriteLine();
    }
}

Console.WriteLine(" –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø –†–ê–ë–û–¢–´ –° –ñ–ò–í–û–¢–ù–´–ú–ò ");
Console.WriteLine();

IAnimal dog = new Dog("–ë–æ–±–∏–∫", 3, "–õ–∞–±—Ä–∞–¥–æ—Ä");
IAnimal cat = new Cat("–ú—É—Ä–∫–∞", 2, true);
IAnimal bird = new Bird("–ö–µ—à–∞", 1, 15.5);
IAnimal fish = new Fish("–ù–µ–º–æ", 1, "–º–æ—Ä—Å–∫–∞—è");

AnimalService dogService = new AnimalService(dog);
AnimalService catService = new AnimalService(cat);
AnimalService birdService = new AnimalService(bird);
AnimalService fishService = new AnimalService(fish);

dogService.DisplayAnimalInfo();
catService.DisplayAnimalInfo();
birdService.DisplayAnimalInfo();
fishService.DisplayAnimalInfo();

Console.WriteLine(" –ï–ñ–ï–î–ù–ï–í–ù–ê–Ø –†–£–¢–ò–ù–ê –ñ–ò–í–û–¢–ù–´–•:");
dogService.PerformDailyRoutine();
catService.PerformDailyRoutine();
birdService.PerformDailyRoutine();
fishService.PerformDailyRoutine();

Console.WriteLine("üêï‚Äçü¶∫ –†–ê–ë–û–¢–ê –° –ö–û–õ–õ–ï–ö–¶–ò–ï–ô –ñ–ò–í–û–¢–ù–´–•:");
List<IAnimal> animals = new List<IAnimal> { dog, cat, bird, fish };

foreach (var animal in animals)
{
    var service = new AnimalService(animal);
    service.DisplayAnimalInfo();
}
````
